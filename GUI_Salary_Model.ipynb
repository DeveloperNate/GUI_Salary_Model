```python
import tkinter as tk
import tkinter.ttk as ttk
from PIL import ImageTk, Image
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.externals import joblib
    

img = Image.open(r"C:\Users\NJGaj\port\Tkinter\background2.jpg")

class SampleApp(tk.Frame): #change to tk.Frame
    def __init__(self, parent, *args, **kwargs): #added parent
    
        tk.Frame.__init__(self, parent, *args, **kwargs) #added parent
        

        # the container is where we'll stack a bunch of frames
        # on top of each other, then the one we want visible
        # will be raised above the others
        container = tk.Frame(parent) #changed self to parent
        self.bk_ground = ImageTk.PhotoImage(img) # creates the background image for all windows
        self.w = self.bk_ground.width() 
        self.h = self.bk_ground.height()
    
        container.pack(side="top", fill="both", expand=True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)
        

        self.frames = {}
        for F in (StartPage, PageOne , AnswerPage):
            page_name = F.__name__
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame

            # put all of the pages in the same location;
            # the one on the top of the stacking order
            # will be the one that is visible.
            frame.grid(row=0, column=0, sticky="nsew")
        
        self.show_frame("StartPage")

    def show_frame(self, page_name):
        # function that switches between window
        frame = self.frames[page_name]
        frame.tkraise()
        
    def show_answer_frame(self,page_name,prediction,complete_data):
        # Function that switched to the final window after modelling has been completed
        frame = self.frames[page_name]
        frame.tkraise()
        frame.AnswerDisplay(prediction,complete_data)

class StartPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        # create canvas to link image and frame
        self.bk = tk.Canvas(self,width=self.controller.w, height=self.controller.h)
        self.bk.pack()
        self.bk.create_image(0, 0, image=self.controller.bk_ground, anchor='nw')
        
        # create labels, buttons in page
        self.canvas_title = self.bk.create_text(400, 10, anchor="n", fill="white", text="Salary Predictor", font=("arial", 18))
        self.canvas_desciption = self.bk.create_text(400, 50, anchor="n", fill="white", text="Enter your information and predict your salary", font=("arial", 14))
        self.canvas_button_quit = tk.Button(self.bk,text = "Quit", command = root.destroy, anchor = "w")
        self.canvas_button_start = tk.Button(self.bk,text = "Enter Your Information",command=lambda: controller.show_frame("PageOne"), anchor = "w")
        self.button1_window = self.bk.create_window(390, 200, anchor="nw", window=self.canvas_button_quit)
        self.button2_window = self.bk.create_window(340, 170, anchor="nw", window=self.canvas_button_start)
       
    


class PageOne(tk.Frame):
    
    # Lists needed for the dropdown boxes
    
    work_list = ['Federal Gov', 'Local Gov', 'Private',
                     'Self Employed Inc', 'Self Employed not Inc','State Gov', 'Unemployed']
        
    education_list = ['Assoc Acdm', 'Assoc Voc', 'Bachelors',
                          'Doctorate', 'Dropout', 'HighSchool Grad','Masters', 'Prof-school','College']
        
    Native_list = ['First World Country','Second World Country','Third World Country', 'United-States']
        
    occupation_list = ['Admin', 'Armed Forces','Craft-repair', 'Managerial Position',
                           'Farming/Fishing', 'Cleaners','Engineer', 'Other Service',
                           'Private House Service', 'Prof Specialty','Protective Service', 'Sales','Tech Support', 'Transport']
        
    married_status_list = ['Married','Unmarried','Own-child','Other-relative']
    relationship_list = ['Married','Never-married','Divorced','Widowed','Married-spouse-absent']
    race_list = ['Asian','Black','Other','White']
    gender_list = ['Female', 'Male']
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
    
        # Create Canvas that links with image 
        self.bk1 = tk.Canvas(self,width=self.controller.w, height=self.controller.h)
        self.bk1.pack()
        self.bk1.create_image(0, 0, image=self.controller.bk_ground, anchor='nw')
        
        #Create Title
        self.canvas_title1 = self.bk1.create_text(400, 10, anchor="n", fill="white", text="Enter Your Information", font=("arial",18))
        
        # Lists needed for the dropdown boxes
    
        work_list = ['Federal Gov', 'Local Gov', 'Private',
                     'Self Employed Inc', 'Self Employed not Inc','State Gov', 'Unemployed']
        
        education_list = ['Assoc Acdm', 'Assoc Voc', 'Bachelors',
                          'Doctorate', 'Dropout', 'HighSchool Grad','Masters', 'Prof-school','College']
        
        Native_list = ['First World Country','Second World Country','Third World Country', 'United-States']
        
        occupation_list = ['Admin', 'Armed Forces','Craft-repair', 'Managerial Position',
                           'Farming/Fishing', 'Cleaners','Engineer', 'Other Service',
                           'Private House Service', 'Prof Specialty','Protective Service', 'Sales','Tech Support', 'Transport']
        
        married_status_list = ['Married','Unmarried','Own-child','Other-relative']
        relationship_list = ['Married','Never-married','Divorced','Widowed','Married-spouse-absent']
        race_list = ['Asian','Black','Other','White']
        gender_list = ['Female', 'Male']
        
        # dropdown forms
        self.Native  = ttk.Combobox(self.bk1, values=Native_list, width=15)
        self.work= ttk.Combobox(self.bk1, values = work_list,width=15)
        self.Occupation = ttk.Combobox(self.bk1,values = occupation_list, width=15)
        self.education = ttk.Combobox(self.bk1,values = education_list,width =15)
        self.married_status = ttk.Combobox(self.bk1, values=married_status_list,width = 15)
        self.relationship = ttk.Combobox(self.bk1, values= relationship_list,width = 15)
        self.race = ttk.Combobox(self.bk1,values= race_list, width =15 )
        self.gender = ttk.Combobox(self.bk1, values = gender_list , width = 15)
        
        
        #Other input forms 
        self.age = tk.Entry(self.bk1, width=18)
        self.hours_worked = tk.Entry(self.bk1,width=18)
        self.captial_gain = tk.Entry(self.bk1,width =18)
        self.captial_loss = tk.Entry(self.bk1,width =18)
        self.captial_gain.insert(0,'0')
        self.captial_loss.insert(0,'0')
        
        #button
        self.canvas_button_start1 = tk.Button(self.bk1,text = "Home",command=lambda: controller.show_frame("StartPage"), anchor = "w")
        self.canvas_enter = tk.Button(self.bk1,text = "Enter",command =  self.get_data, anchor = "w")
        
        #create windows to combine canvas with entry boxes to create forms
        self.native_drop_down = self.bk1.create_window(200,150,anchor="nw", window=self.Native)
        self.work_drop_down = self.bk1.create_window(580,150,anchor="nw", window=self.work)
        self.occupation_list = self.bk1.create_window(200,180, anchor='nw', window = self.Occupation)
        self.education_win = self.bk1.create_window(580,180, anchor ='nw',window=self.education)
        self.married = self.bk1.create_window(200,210, anchor='nw', window = self.married_status)
        self.relat = self.bk1.create_window(580,210, anchor ='nw',window=self.relationship)
        self.age_win = self.bk1.create_window(200,240,anchor ='nw',window=self.age)
        self.race_win = self.bk1.create_window(580,240, anchor='nw', window = self.race)
        self.gender_win = self.bk1.create_window(200,270, anchor='nw',window = self.gender)
        self.hours_worked_win = self.bk1.create_window(580,270, anchor='nw',window = self.hours_worked)
        self.captial_gain_win = self.bk1.create_window(200,300, anchor = 'nw', window = self.captial_gain)
        self.captial_loss_win = self.bk1.create_window(580,300, anchor = 'nw', window = self.captial_loss)
        
        # combine buttons 
        self.button2_window1 = self.bk1.create_window(340, 330, anchor="nw", window=self.canvas_button_start1)
        self.button2_enter = self.bk1.create_window(450, 330, anchor="nw", window=self.canvas_enter)
        
        # combine labels 
        self.infobox = self.bk1.create_text(280,100,anchor="nw", fill="white", text="Complete all the fields", font=("arial", 12))
        self.lb_native = self.bk1.create_text(20, 150, anchor="nw", fill="white", text="Where do you descent from?", font=("arial", 10))
        self.lb_sector = self.bk1.create_text(400, 150, anchor="nw", fill="white", text="What sector do you work in? ", font=("arial", 10))
        self.lb_occupation = self.bk1.create_text(20, 180, anchor="nw", fill="white", text="What is your job?", font=("arial", 10))
        self.lb_eductation = self.bk1.create_text(400, 180, anchor="nw", fill="white", text="What is your education? ", font=("arial", 10))
        self.lb_married = self.bk1.create_text(20, 210, anchor="nw", fill="white", text="What is your main relationship?", font=("arial", 10))
        self.lb_relation = self.bk1.create_text(400, 210, anchor="nw", fill="white", text="What is your martial status?", font=("arial", 10))
        self.lb_age = self.bk1.create_text(20,240, anchor ='nw', fill='white',text='What is your age?', font=('arial',10))
        self.lb_race = self.bk1.create_text(400,240, anchor='nw', fill='white',text='What is your ethnicity?', font=('arial',10))
        self.lb_gender = self.bk1.create_text(20,270, anchor='nw', fill='white',text='What is your Gender?' ,font=('arial',10))
        self.lb_hours = self.bk1.create_text(400,270,anchor='nw',fill='white',text='How many hours do you work?', font=('arial',10))
        self.lb_captial_gain = self.bk1.create_text(20,300, anchor='nw', fill='white',text='What is your Capital gain?' ,font=('arial',10))
        self.lb_captial_loss = self.bk1.create_text(400,300,anchor='nw',fill='white',text='What is your Capital loss?', font=('arial',10))
        
    def get_data(self):
            # Simple Validation - checks to ensure all fields have been entered                                                                                                              
            nativeinfo = self.Native.get()
            workinfo = self.work.get()
            occupationinfo = self.Occupation.get()
            married_status_info = self.married_status.get()
            relationship_info = self.relationship.get()
            race_info = self.race.get()
            gender_info=self.gender.get()
            ageinfo = self.age.get()
            hours_info=self.hours_worked.get()
            capital_gain_info = self.captial_gain.get()
            capital_loss_info = self.captial_loss.get()
            education_info = self.education.get()
            all_info = [nativeinfo,workinfo ,occupationinfo ,married_status_info ,
                        relationship_info ,race_info ,gender_info ,ageinfo, hours_info ,
                        capital_gain_info ,capital_loss_info ,education_info]
        
            look_for = ""
           
            if look_for in all_info:
                self.bk1.itemconfig(self.infobox, text="Please complete the required field!")
            else:
                self.norm_data(nativeinfo,workinfo,occupationinfo,married_status_info,relationship_info,race_info,gender_info,ageinfo,hours_info,capital_gain_info,capital_loss_info,education_info)

                
    def norm_data(self,nativeinfo,workinfo,occupationinfo,married_status_info,relationship_info,race_info,gender_info,
                  ageinfo,hours_info,capital_gain_info,capital_loss_info,education_info):
        #Normalises data and use model to predict salary classification
                                                                                                                          
                                                                                                                          
        # list of all possible options                                                                                             
        work_list = ['Federal Gov', 'Local Gov', 'Private',
                     'Self Employed Inc', 'Self Employed not Inc','State Gov', 'Unemployed']
    
        education_list = ['Assoc Acdm', 'Assoc Voc', 'Bachelors',
                          'Doctorate', 'Dropout', 'HighSchool Grad','Masters', 'Prof-school','College']
    
        marital_list = ['Divorced','Married', 'Married but Spouse Absent','Never Married', 'Widowed']
    
        occupation_list = ['Admin', 'Armed Forces','Craft-repair', 'Managerial Position',
                           'Farming/Fishing', 'Cleaners','Engineer', 'Other Service',
                           'Private House Service', 'Prof Specialty','Protective Service', 'Sales',
                           'Tech Support', 'Transport']
    
        relationship_list = ['Married', 'Other Relative','Own Child', 'Unmarried']
    
        race_list = ['American Indian/Eskimo', 'Asian Pac-Islander', 'Black','Other', 'White']
    
        native_list = ['First World Country','Second World Country','Third World Country', 'United-States']
    
        gender_list =['Female', 'Male']
        
        
        # creates list for the loop                                                                                                                  #
        loop_list=[work_list,education_list,
               marital_list,occupation_list,relationship_list,race_list,gender_list,native_list]
    
        info =[workinfo,education_info,married_status_info,occupationinfo,relationship_info,race_info,gender_info,nativeinfo] 
    
        array_dropdown = []
    
        # changes the data into 1 or 0 so that it can be put into the model
        for counter,i in enumerate(loop_list):
            for x in i :
                if x == info[counter] :
                    array_dropdown.append(1)
                else:
                    array_dropdown.append(0)
        
        # gathers all data that needs to scaled to value between 0 and 1                                                                                                                 
        array_age = ageinfo
        array_captial_gain = capital_gain_info
        array_captial_loss = capital_loss_info
        array_hours = hours_info                                                                                                                    
        normal_data = [array_age,array_captial_gain,array_captial_loss,array_hours]
                                                                                                                          
        normal_data = np.reshape(normal_data, (-1,1)) # reshape array to fit into scaler
        
        #loads scaler and normaise data into a list                                                                                                                  
        pickle_file = r"C:\Users\NJGaj\port\Scikit-Learn\K Nearest Neighbors\Salary\MinMaxScaler.pkl"
        scaler = joblib.load(pickle_file)                                                                                                                                                                                                                                 
        scaled_data = scaler.fit_transform(normal_data)
        scaled_data = [item for sublist in scaled_data for item in sublist]    
    
        # combines and reshapes data to be used model predictor                                                                                                                
        complete_data = scaled_data + array_dropdown
        complete_data = np.reshape(complete_data, (1,-1))                                                                                                                  
                                                                                                                          
        # loads model and uses data to make prediction
        clf_file = r"C:\Users\NJGaj\port\Scikit-Learn\K Nearest Neighbors\Salary\rf_classifier_salary.pkl"
        clf = joblib.load(clf_file)
        prediction = clf.predict(complete_data)
        self.controller.show_answer_frame("AnswerPage",prediction,complete_data)
            
        
class AnswerPage(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        # creates canvas and links with image                                                                                                                  
        self.bk2 = tk.Canvas(self,width=self.controller.w, height=self.controller.h)
        self.bk2.pack()
        self.bk2.create_image(0, 0, image=self.controller.bk_ground, anchor='nw')
        #add title to canvas
        self.canvas_title2 = self.bk2.create_text(400, 10, anchor="n", fill="white", text="Based on the information, the salary is .....", font=("arial", 18))
        
    def AnswerDisplay(self,answer,complete_data):
        # display classifcation and shows buttons to clarify if the data is correct                                                                                                                  
        if answer == 1 :
            self.canvas_answer = self.bk2.create_text(400, 60, anchor="n", fill="white", text="Above 50,000", font=("arial", 18))
        elif answer == 0 :
            self.canvas_answer = self.bk2.create_text(400, 60, anchor="n", fill="white", text="Below 50,000", font=("arial", 18))
        else:
            self.canvas_answer = self.bk2.create_text(400, 60, anchor="n", fill="white", text="Error", font=("arial", 18))
        
        # The correct label if wrong
        label = int(1 - answer)
        self.correct_button = tk.Button(self.bk2, text=" Correct" , command = lambda : self.add_file(complete_data,answer))
        self.incorrect_button = tk.Button(self.bk2, text=" Incorrect" , command = lambda : self.add_file(complete_data,label))
        
        correct_button_win = self.bk2.create_window(320, 150, anchor="nw", window=self.correct_button)
        incorrect_button_win = self.bk2.create_window(420, 150, anchor="nw", window=self.incorrect_button)
        
        
    def add_file(self,complete_data ,correct):
        # records the new data and classification to a file to be used to improve model 
    
        complete_data= [item for sublist in complete_data for item in sublist]
        complete_data = complete_data.insert([0],correct)
    
        f=open("salary_data.txt", "a+")
        length = len(complete_data)
        for counter,data in enumerate(complete_data):
            if counter < length  :
                f.write("{},".format(data))
            elif counter == length :
                f.write("{} \n".format(data))
            else:
                f.write("error")
        self.controller.show_frame("StartPage")
                
        
        
if __name__ == "__main__":
    root = tk.Tk() #Added
    app = SampleApp(root) #added root
    root.mainloop()
```
